```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

Generate the data based on **y=mx+c** equation

```python
def generate_data(n, slope, constant):
  X = []
  y = []
  for i in range(n):
      X.append(i)
      target = slope*i + constant;
      y.append(target)
  return X, y
```

```python
X, y = generate_data(11, 2, 1);
```

```python
X, y
```

Convert data to numpy arrays

```python
x= np.array(X, dtype= float)
y= np.array(y, dtype= float)
```

Make simple sequential model with "sgd" as optimizer and loss as "mse"

```python
model=Sequential([Dense(units=1, input_shape=[1])])
model.compile(optimizer="sgd", loss="mean_squared_error", metrics=['acc'])
```

```python
model.fit(x,y, epochs=500)
```

```python
print(model.predict([10]))
```
