```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras


train = pd.read_csv("/Users/vikaspuri/ArtificialIntelligence/MachineLearningApplications/Titanic/train.csv")
test = pd.read_csv("/Users/vikaspuri/ArtificialIntelligence/MachineLearningApplications/Titanic/test.csv")
```

```python
train = train.rename(columns = {'Pclass' : 'TicketClass'})
test = test.rename(columns = {'Pclass' : 'TicketClass'})
```

```python
train = train.drop(['Name','Ticket','Fare','Cabin','Embarked'],axis =1)
test = test.drop(['Name','Ticket','Fare','Cabin','Embarked'],axis =1)
```

```python
test
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_sex = LabelEncoder()
train.iloc[:,3]  = label_encoder_sex.fit_transform(train.iloc[:,3])
test.iloc[:,2] = label_encoder_sex.fit_transform(test.iloc[:,2])
```

```python
train
```

```python
train = train[['PassengerId','Sex','TicketClass','SibSp','Parch','Survived']]
test = test[['PassengerId','Sex','TicketClass','SibSp','Parch']]
```

```python
train
```

```python
X_train = train.iloc[:,0:5]   
y_train = train.iloc[:,5]  
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler();
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(3, activation='relu', kernel_initializer='glorot_uniform', input_dim = 5))
classifier.add(Dense(2, activation='relu', kernel_initializer='glorot_uniform'))
classifier.add(Dense(1, activation='sigmoid'))
             
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochsprediction = classifier.predict(X_test).tolist() = 100)
```

```python
prediction = classifier.predict(X_test).tolist()
```
